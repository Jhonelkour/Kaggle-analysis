##### 9-Find the mean and standard deviation of height using the simple operation only 

```python
data
```

```python
Mean_height = Height.sum()/Height.shape[0]

print("The mean of height with simple operation : ", Mean_height)
```

```python
squared_differences = (Height - Mean_height) ** 2
variance = np.sum(squared_differences)/Height.shape[0]
std_dev = np.sqrt(variance)
print("Standard Deviation:", std_dev)

```
